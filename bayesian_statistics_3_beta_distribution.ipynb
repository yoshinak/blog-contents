```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

表が 50 回、裏が 50 回出たとした場合の事前分布を仮定します。

コインの表、裏が 50 回ずつ出る場合を想定した事前確率の確率密度関数グラフです。

表 ( $ m $ ) 、裏 ( $ n $ ) の出る回数を 50 で設定します。

```python
m, n = 50, 50
a, b = m + 1, n + 1
```

横軸に確率変数として、確率 $ p $ を設定します。

```python
p = np.linspace(0.01, 0.99, 100)
```

縦軸が確率密度 ( 確率 $ p $ の相対的な出やすさ ) のグラフを描きます。

```python
plt.figure(figsize=(10,10))
plt.xlim(0.0, 1)
plt.plot(p, beta.pdf(p, a, b), 'b-')
plt.xlabel('p', fontsize='15')
plt.ylabel('P^{pre}(p)', fontsize='15')
plt.show()
```
